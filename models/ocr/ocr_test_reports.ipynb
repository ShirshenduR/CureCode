```python
import cv2
import pytesseract
from PIL import Image
import numpy as np
```

```python
# Path to Tesseract executable (for Windows users)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Function to display images in Jupyter
import IPython.display as display
def show_image(image, title="Image"):
    display.display(Image.fromarray(image))
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Grayscale
    thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)[1]  # Binarization
    denoised = cv2.medianBlur(thresh, 3)  # Denoising
    return denoised
```

```python
image_path = "C:/Users/Dhairya/Documents/GitHub/CureCode/dataset/page_2.jpg"
processed_img = preprocess_image(image_path)
show_image(processed_img, "Preprocessed Image")
```

```python
extracted_text = pytesseract.image_to_string(image_path)
print("Extracted Text:\n", extracted_text)
```
